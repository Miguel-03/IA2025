```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("/content/Salary.xlsx")
data.head()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
x=df[['YearsExperience']]
y=df[['Salary']]

plt.figure(figsize=(16,8))
plt.scatter(x,y, color= "pink", label="Datos Reales")
plt.title("Salario vs Años de Experiencia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data.columns
X=data['YearsExperience']
Y=data['Salary']
model=LinearRegression()
model.fit(X.values.reshape(-1,1),Y)
y_pred=model.predict(X.values.reshape(-1,1))
r2=r2_score(Y,y_pred)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, color="pink", label="Datos Reales")
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
plt.plot(x,y_pred, color="#76b5c5", linewidth=2, label="regresion Lineal")
plt.title("Salario vs Años de Experiencia")
plt.ylabel("Salario")
plt.xlabel("Años de Experiencia")
plt.legend()
plt.grid(True)
plt.show()

r2=r2_score(y,y_pred)
print(f"Ecuacion de la recta y= {model.coef_[0]:2f}x+ {model.intercept_:2f}")
print(f"Coeficiente de Determinación (R2): {r2:2f}")
```
